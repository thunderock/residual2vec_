```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from debias_graph import debias_wrapper
from we import doPCA
from we_utils import get_direction
from utils import graph_utils
from models import fast_knn_cpu
```

```python
dataset = "polbook"
# group values for each node
y = pd.read_csv("../../final_128/{}/node_table.csv".format(dataset)).group_id.values
y
```

```python
def plot_embs(emb, y, pallete=None, method='PCA', title="no_idea"):
    if not pallete:
        pallete = {0: "blue", 1: "red", 2: "purple"}
    n = y.shape[0]
    if method == "PCA":
        xy = PCA(n_components=2).fit_transform(emb)
    else:
        pass
    
    df = pd.DataFrame({
        "x": xy[:, 0],
        "y": xy[:,1],
        "hue": y,
    })
    sns.scatterplot(df, x="x", y="y", hue="hue", palette=pallete )
    plt.title(title)
```

```python
# load 
embs = {}
embs['deepwalk'] = np.load("../../final_128/{}/{}_one/{}_deepwalk.npy".format(dataset, dataset, dataset))
```

```python
# definitional words, these are supposed to be represent the group,
# in this case lets take these to be the nodes closest to centroid of group
# in this case are the centroids of the groups

centroids = graph_utils.get_centroid_per_group(embs['deepwalk'], y)
definitional = graph_utils.get_n_nearest_neighbors_for_nodes(
    nodes=centroids, 
    embs=embs['deepwalk'],
    k=1,
    metric='cosine'
)

# create pairs
dim = 128
defn = np.zeros((3, 2))

defn[0, 0] = definitional[1]
defn[0, 1] = definitional[2]

defn[1, 0] = definitional[0]
defn[1, 1] = definitional[2]

defn[2, 0] = definitional[0]
defn[2, 1] = definitional[1]
```

```python
# gender specific nodes, these are used to figure out which nodes should be normalized, 
# in paper "neutralization", in our case picking ones which have closest to these centroid 
# for each group

gender_specific_nodes = graph_utils.get_n_nearest_neighbors_for_nodes(
    nodes=centroids, 
    embs=embs['deepwalk'],
    k=10,
    metric='cosine'
)
```

```python
# equalize words, these words should be equidistant from gender plane, in our case 
# these are most distant pairs of different classes 
equalize = graph_utils.get_farthest_pairs(embs['deepwalk'], y, same_class=False, per_class_count=1)
```

```python
definitional.shape, equalize.shape, gender_specific_nodes.shape
```

```python
# drection from LDA from base embeddings
direction = get_direction(embs['deepwalk'], y, "LDA")
embs['LDA_not_changing_gender_specific'] = debias_wrapper(emb=embs['deepwalk'], gender_specific_words=gender_specific_nodes, 
               definitional=defn, equalize=equalize, y=y, direction=direction,
                          drop_gender_specific_words=False)
embs['LDA_changing_gender_specific'] = debias_wrapper(emb=embs['deepwalk'], gender_specific_words=gender_specific_nodes, 
               definitional=defn, equalize=equalize, y=y, direction=direction,
                          drop_gender_specific_words=True)
```

```python
# drection from PCA from base embeddings
direction = get_direction(embs['deepwalk'], y, "PCA")

embs['PCA_not_changing_gender_specific'] = debias_wrapper(emb=embs['deepwalk'], gender_specific_words=gender_specific_nodes, 
               definitional=defn, equalize=equalize, y=y, direction=direction,
                          drop_gender_specific_words=False)
embs['PCA_changing_gender_specific'] = debias_wrapper(emb=embs['deepwalk'], gender_specific_words=gender_specific_nodes, 
               definitional=defn, equalize=equalize, y=y, direction=direction,
                          drop_gender_specific_words=True)
```

```python
# taking direction from definitional and embedding, not changing gender specific words 
# as in paper
direction = doPCA(defn, embs['deepwalk'], num_components=1).components_[0]
embs['original_paper'] = debias_wrapper(emb=embs['deepwalk'], gender_specific_words=gender_specific_nodes, 
               definitional=defn, equalize=equalize, y=y, direction=direction,
                          drop_gender_specific_words=False)
```

```python
data = []

for idx, (k, emb) in enumerate(embs.items()):
    xy = PCA(n_components=2).fit_transform(emb)
    data.append(
        pd.DataFrame({
            "x": xy[:, 0],
            "y": xy[:,1],
            "hue": y,
            "model": k
        })
               )
df = pd.concat(data)
```

```python
sns.set_style("white")
sns.set(font_scale=1)
sns.set_style("ticks")
# fig, ax = plt.subplots(figsize=(7,5))

g = sns.FacetGrid(
    data=df,
#     row="row",
    col='model',
    sharex=False,
    sharey=False,
    col_wrap=3
    
)

g.map(sns.scatterplot, "x", "y", "hue", sizes=(10, 200), palette ={0: "blue", 1: "red", 2: "purple"}, )
for (i,j,k), data in g.facet_data():

    ax = g.facet_axis(i, j)
    # need to see a better way of removing title
    title = ax.get_title()
    ax.set_title(title.replace("model =", ""))
```
